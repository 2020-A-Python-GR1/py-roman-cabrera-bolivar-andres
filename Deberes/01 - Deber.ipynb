```python
import csv
import sys

def insert_dataitems(data):
    with open("items.csv", "a+") as csv_file:   
        writer = csv.writer(csv_file)
        writer.writerows(data)

def insert_rowfac(row):
    with open("facturas.csv", "a+") as csv_file:   
        writer = csv.writer(csv_file)
        writer.writerow(row)
        

def get_last_numfac():
    data = []
    with open("facturas.csv", "r") as csv_file:
        reader = csv_file.readlines()
        for row in reader:
            if not row[0] == '\n':
                data.append(row[0])
    if data:
        return int(data[-1])
    else:
        return 0

def get_rowfac_by_index(index):
    with open("facturas.csv", "r") as csv_file:
        reader = csv.reader(csv_file)
        for row in reader:
            if row:
                if int(row[0]) == index:
                    return row
                
def get_items_by_facindex(index):
    data = []
    with open("items.csv", "r") as csv_file:
        reader = csv.reader(csv_file)
        for row in reader:
            if row:
                if int(row[0]) == index:
                    data.append(row)
    return data


def del_rowfac_by_index(index):
    data = list()
    with open("facturas.csv", "r") as csv_file:
        records = csv_file.readlines()
        for record in records:
            if not record == '\n':
                data.append(record)
                
    with open("facturas.csv", "w") as csv_file:
        for row in data:
            if not row.split(",")[0] == index:
                csv_file.write(row)
                csv_file.write('\n')


def del_rowitem_by_index(index):
    data = list()
    with open("items.csv", "r") as csv_file:
        records = csv_file.readlines()
        for record in records:
            if not record == '\n':
                data.append(record)
                
    with open("items.csv", "w") as csv_file:
        for row in data:
            if not row.split(",")[0] == index:
                csv_file.write(row)
                csv_file.write('\n')                
                

        
def insertar_nueva_factura():
    print("Ingresar datos de factura")
    nombre_fac = input("Nombre: ")
    fecha_fac = input("Fecha: ")
    num_fac = get_last_numfac() + 1
    
    with open("facturas.csv", "a+") as csv_file:   
        writer = csv.writer(csv_file)
        writer.writerow([num_fac, nombre_fac, fecha_fac])
        
    while True:
        insertar_item_factura(num_fac)
        response = input("Ingresar otro Ã­tem?:  [Y/N] ")
        if response == "N":
            break
 
def insertar_item_factura(num_fac):
    print("Ingresar items de factura")
    cantidad_item = input("Cantidad: ")
    descr_item = input("Descripcion: ")
    precio_item = input("Precio: ")
    
    with open("items.csv", "a+") as csv_file:
        writer = csv.writer(csv_file)
        writer.writerow([num_fac, cantidad_item, descr_item, precio_item])
        

        
def consultar_fac():
    print("Consulta de facturas")
    num_fac = input("Ingrese numero de factura: ")
    fac_data = get_rowfac_by_index(int(num_fac))
    if fac_data:
        print("\n----------Datos Factura----------")
        print(f"Numero factura: {fac_data[0]}")
        print(f"Nombre: {fac_data[1]}")
        print(f"Fecha: {fac_data[2]}")
        print("\n----------Items Factura----------")
        item_data = get_items_by_facindex(int(num_fac))
        print("Cantidad\tDescripcion\tPrecio")
        for item in item_data:
            print(f"{item[1]}\t\t{item[2]}\t\t{item[3]}")
    else:
        print("No encontrado")
        

        
def update_fac_data():
    print("Actualizacion de facturas")
    num_fac = input("Ingrese numero de factura: ")
    fac_data = get_rowfac_by_index(int(num_fac))
    if fac_data:
        del_rowfac_by_index(num_fac)
        print("----------Datos actuales----------")
        print(f"Numero factura: {fac_data[0]}")
        print(f"Nombre: {fac_data[1]}")
        print(f"Fecha: {fac_data[2]}")
        
        print("\nIngresar nuevos datos de factura")
        dec = input("Ingrese parametro a modificar: \n1.Nombre\n2.Fecha\n")
        
        if int(dec) == 1:
            new_nombre_fac = input("Nombre: ")
            fac_data[1] = new_nombre_fac
            
        if int(dec) == 2:
            new_fecha_fac = input("Fecha: ")
            fac_data[2] = new_fecha_fac
        
        insert_rowfac(fac_data)
        dec_up = input("Actualizar items? [Y/N]")
        if dec_up == "Y":
            update_item_data(num_fac)       
    else:
        print("No encontrado")
        
        
        
        
def update_item_data(num_fac):
    data = get_items_by_facindex(int(num_fac))
    del_rowitem_by_index(str(num_fac))
    desc = input("Ingrese nombre de item a modificar")
    for item in data:
        if item[2] == desc:
            dec = input("Ingrese parametro a modificar: \n1. Cantidad\n2. Descripcion\n3. Precio\n")
            if int(dec) == 1:
                new_cantidad = input("Cantidad: ")
                item[1] = new_cantidad
            if int(dec) == 2:
                new_desc = input("Descripcion: ")
                item[2] = new_desc
            if int(dec) == 3:
                new_price = input("Precio: ")
                item[3] = new_price
                
    insert_dataitems(data)

    
    
def main():
    print("Sistema CRUD - Facturas e items")
    dec = input("Escoja una opcion \n1. Ingresar factura\n2. Consultar factura\n3. Actualizar factura\n4. Eliminar factura\n")
    if dec == "1":
        insertar_nueva_factura()
    elif dec == "2":
        consultar_fac()
    elif dec == "3":
        update_fac_data()
    elif dec == "4":
        dec_del = input("Ingrese numero de factura a eliminar")
        del_rowfac_by_index(dec_del)
        del_rowitem_by_index(dec_del)
    else:
        print("Ingrese una opcion valida")
        
        
                
    

if __name__ == "__main__":
    main()
```
