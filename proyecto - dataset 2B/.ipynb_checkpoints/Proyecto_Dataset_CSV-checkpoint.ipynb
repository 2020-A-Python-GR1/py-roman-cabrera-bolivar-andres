En este proyecto se ha seleccionado un dataset que contiene información sobre marcas de laptops de diferentes fabricantes que tiene los siguientes parámetros

* Company: nombre del frabricante
* Product: nombre del producto
* TypeName: clasificación del producto (laptop, notebook, ultrabook, etc.)
* Inches: tamaño de la pantalla
* ScreenResolution: resolución de la pantalla
* Cpu: procesador del equipo
* Ram: memoria RAM en GB
* Memory: disco duro del equipo
* Gpu: chipset gráfico del equipo
* OpSys: sistema operativo por defecto
* Weight: peso del equipo
* Price_euros: precio en euros

El conjunto de datos contiene aproximadamente tiene 1300 instancias aproximadamente. Fue extraído de Kaggle con el siguiente [enlace](https://www.kaggle.com/ionaskel/laptop-prices)


## Carga de datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

path = r'laptops.csv'

columns = ['Company','Product','TypeName','Inches','ScreenResolution','Cpu','Ram','Memory','Gpu','OpSys','Weight','Price_euros']
df = pd.read_csv(path, usecols = columns)
```

### 1) Porcentaje de equipos según su tipo
Agrupación de equipos según la categoría a la que pertenecen

```python
brand = df['TypeName'].value_counts()
fig_1 = plt.figure(figsize=(10,10))
plt.pie(brand, labels= brand.index, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

### 2) Top 10 de marcas con más productos
Agrupación de productos según su marca

```python
device_type = df['Company'].value_counts().head(10)
fig_2 = plt.figure(figsize=(10,10))
plt.pie(device_type, labels= device_type.index, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

### 3) Top 10 de CPUs más usados
Agrupación por los CPUs más usados en los dispositivos

```python
cpu = df['Cpu'].value_counts().head(10)
fig_3 = plt.figure(figsize=(10,10))
plt.pie(cpu, labels= cpu.index, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

### 4) Equipos por intervalo de pesos en kilogramos
Agrupación de equipos según el intervalo de peso al que pertenecen

```python
df['Weight'] = df['Weight'].map(lambda x: x.strip('kg'))
weight = pd.cut(pd.to_numeric(df['Weight']), bins=[0, 1 ,1.5 ,2 ,2.5 ,3 ,3.5 ,4])
ax = weight.value_counts(sort=False).plot.bar(rot=0, figsize=(20,5))
plt.show()
```

### 5) Productos agrupados por tamaño de memoria RAM
Productos agrupados por la memoria que tienen los dispositivos

```python
df['Ram'] = df['Ram'].map(lambda x: x.strip('GB'))
ram = pd.cut(pd.to_numeric(df['Ram']), bins=[0, 4 ,8 ,16 ,32])
ax = ram.value_counts(sort=False).plot.bar(rot=0, figsize=(20,5))
plt.show()
```

### 6) Top 10 de GPUs más usadas
Agrupación por las GPUs más usadas en los dispositivos

```python
gpu = df['Gpu'].value_counts().head(10)
fig_6 = plt.figure(figsize=(10,10))
plt.pie(gpu, labels= gpu.index, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

### 7) Top de productos fabricados por Dell
Agrupación de productos por marca y por tipo de producto

```python
dell = df.loc[df['Company'] == 'Dell'].groupby(by='TypeName')['TypeName']
fig_7 = plt.figure(figsize=(10,10))
plt.bar(dell.count().index , dell.count().values)
plt.xticks(rotation=90)
plt.show()
```

### 16) Variaciones de precios de los diferentes fabricantes
Variaciones de precio linealmente según todos los productos

```python
prices = df['Price_euros'].sort_values(ascending=False).head(50)
names = df['Product'].iloc[prices.index]
fig_16 = plt.figure(figsize=(20,5))
plt.plot(prices.values , names.values)
plt.xticks(rotation=90)
plt.show()
```
